# Intern marks visualization

```python
import matplotlib.pyplot as plt

# Data for Manish's attempts
manish_dates = ['26-07-2023', '26-07-2023']
manish_scores = [3, 5]
# attempt=['First attempt','Second attempt']

# Data for Amit's attempts
amit_dates = ['26-07-2023', '30-07-2023','01-08-2023']
amit_scores = [3.5, 4.50,5]
# attempt=['First attempt','Second attempt','Third Attempt']

# Plotting Manish's data
plt.plot(manish_dates, manish_scores, label='Manish', marker='o')


# Plotting Amit's data
plt.plot(amit_dates, amit_scores, label='Amit', marker='o')

# Adding labels and title to the plot
plt.xlabel('Date')
plt.ylabel('Scores')
plt.title('Internship Task Scores')
plt.legend()

# Rotating x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot

plt.show()
```

```python
import matplotlib.pyplot as plt

# Manish's scores
scores = [3, 3.5, 5]

# Attempts
attempts = ['First Attempt', 'Second Attempt', 'Third Attempt']

# Plotting the bar chart
plt.bar(attempts, scores, color=['red', 'blue', 'green'],edgecolor='black')
plt.plot(attempts, scores, marker='o', color='purple', linestyle='-', linewidth=2, markersize=2)
plt.scatter(attempts,scores,edgecolor='orange')
plt.xlabel('Attempts')

plt.ylabel('Scores')
plt.title('Manish\'s Scores in Different Attempts')
plt.ylim(0, 5.5)  
plt.show()
```
